### Dicionário de Variáveis
Todas as variáveis que constam neste Dataset foram obtidas apartir dos dados explorados.

 * `iso_cod` - ISO 3166-1 alfa-3 – códigos de país com três letras. Observe que as regiões definidas por OWID (por exemplo, continentes como 'Europa') contêm o prefixo 'OWID_'.
 * `continent` - Continente da localização geográfica.
 * `location`	- Localização geográfica.
 * `date` - Data de observação.
 * `total_cases` - Total de casos confirmados de COVID-19. As contagens podem incluir casos prováveis, quando relatados.
 * `new_cases` - Novos casos confirmados de COVID-19. As contagens podem incluir casos prováveis, quando relatados. Em casos raros em que nossa fonte relata uma alteração diária negativa devido a uma correção de dados, definimos essa métrica como NA.
 * `new_cases_smoothed` - Novos casos confirmados de COVID-19 (suavização de 7 dias). As contagens podem incluir casos prováveis, quando relatados.
 * `total_deaths` - Total de mortes atribuídas à COVID-19. As contagens podem incluir mortes prováveis, quando relatadas.
 * `new_deaths` - Novas mortes atribuídas à COVID-19. As contagens podem incluir mortes prováveis, quando relatadas. Em casos raros em que nossa fonte relata uma alteração diária negativa devido a uma correção de dados, definimos essa métrica como NA. 
 * `new_deaths_smoothed` - Novas mortes atribuídas ao COVID-19 (suavização de 7 dias). As contagens podem incluir mortes prováveis, quando relatadas. 
 * `male_smokers` - Parcela de homens que fumam, ano mais recente disponível.	
 * `handwashing_facilities` - Parcela da população com instalações básicas de lavagem das mãos nas instalações, ano mais recente disponível.
 * `hospital_beds_per_thousand` - Camas hospitalares por 1.000 pessoas, ano mais recente disponível desde 2010.
 * `life_expectancy` - Expectativa de vida ao nascer em 2019.
 * `human_development_index` - Um índice composto que mede o desempenho médio em três dimensões básicas do desenvolvimento humano – uma vida longa e saudável, conhecimento e um padrão de vida decente. Valores para 2019.
 * `population` - População (últimos valores disponíveis).
 * `excess_mortality_cumulative_absolute` - Diferença cumulativa entre o número relatado de mortes desde 1º de janeiro de 2020 e o número projetado de mortes para o mesmo período com base nos anos anteriores.
 * `excess_mortality_cumulative` - Diferença percentual entre o número acumulado de mortes desde 1º de janeiro de 2020 e as mortes projetadas cumulativas para o mesmo período com base nos anos anteriores.
 * `excess_mortality` - Diferença percentual entre o número relatado de mortes semanais ou mensais em 2020–2021 e o número projetado de mortes para o mesmo período com base nos anos anteriores.  
 * `excess_mortality_cumulative_per_millio` - Diferença cumulativa entre o número relatado de mortes desde 1º de janeiro de 2020 e o número projetado de mortes para o mesmo período com base nos anos anteriores, por milhão de pessoas.

#### Análise exploratória de Dados.

Para começarmos iremos importas as bibliotecas necessárias.

```python
# importar as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as pn
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
%matplotlib inline
```

Uma vez que importamos as bibliotécas necessarias, vamos importar o nosso arquivo csv, e antes de iniciarmos qualquer análise iremos ver a cara do nosso Dataset


```python
# importar dataset COVID-19 para um dataframe
df_covid=pd.read_csv("https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv")
# moastrar as 5 primeiras entradas
df_covid.head(5)
```

Feita as importações necessárias, iremos visualizar
 os nomes das nossas colunas

```python
# Visualisando os nomes das colunas
df_covid.columns
```

##  **Tipo de Variáveis e tamanho do nosso Dataframe**
Para desenvolvermos  o nosso projeto precisamos verificar se nossas variáveis estão da forma que precisamos, e também iremos conferir o tamanho do nosso conjuto de Dados.

```python
# Verificar o tipo de variáveis e o tamanho do Dataframe
print("Entradas:\t {}".format(df_covid.shape[0]))
print("Variáveis:\t {}\n".format(df_covid.shape[1]))

# Verificar as 5 primeiras entradas do nosso Dataframe
df_covid.dtypes
```

Analisando o nosso Dataframe verificamos que o tamanho do nosso conjunto de dados é de quase 251.000 entradas e 67 colunas.
Notamos também que a nossa coluna `date`não esté no formato datetime, para prosseguir iremos transforma-la em um formato datetime.

```python
# Transformando a coluna em Data em Datatime
df_covid.date = pd.to_datetime(df_covid.date)
```

### **Verificando os valores ausentes**
A qualidade do nosso conjunto de dados é diretamente influênciada pelos valores ausentes que nele está apresentado.

 Atente-se não para quais dados estão faltando, mas sim o porque desses dados estarem faltando, investigue o motivo.
 


```python
# identificar os valores ausentes por ordem decrescente
(df_covid.isnull().sum() / df_covid.shape[0]).sort_values(ascending=False)
```

#### **Limpeza de dados**
Verificamos que na coluna `location` contém os nomes dos países, e algumas linhas contém os valores agregados de todo o continente, tem também em algumas dessas linhas outras informações.
Como a nossa análise será a nivel de País, resolvi então limpar esses dados.


```python
# Fazendo copia do nosso dataframe
df_clean = df_covid.copy()

# modelando dataframe
df_clean = df_covid[~df_covid['iso_code'].str.contains('OWID_')] 
```

Como alteramos o formato da nosso coluna `date` para o formato datetime, iremos verificar qual a data mais recente do nosso dataset.

```python
# Verificando a data mais recente do DataTime
df_clean.date.max()
```

##### **Dados da Pandemia a nivel de País**
Após termos feitos as ações necessárias em nosso Dataframe, agora iremos passar alguns dados referentes ao Covid-19 no mundo.



```python
# Ver para data atual quais paises tem mais casos.
df_clean.loc[df_covid.date == '2023-01-13'].sort_values(by="total_cases", ascending=False)

```

Nessa extração podemos verificar que os paises que lideram o rank de casos são;

* Estados Unidos - 101,6 milhões de casos
* India - 44,6 milhões de casos
* França - 39,4 milhões de casos
* Alemanha - 37,6 milhões de casos
* Brazil - 36,6 milhões de casos



```python
# Paises com mais mortes
df_clean.loc[df_covid.date == '2023-01-13'].sort_values(by="total_deaths", ascending=False)
```

Após analisarmos os paises que lideram o rank por total de  mortes;


* Estados Unidos - 1.099.823 de mortes	
* Brasil - 695.317 de mortes
* India - 530.726 de mortes 
* Russia - 386.375 de mortes 
* Mexico - 331.510 de mortes

#####**Para ilustar a quantidade de mortes por País, iremos plotar um gráfico de barras**


```python
#paises com mais mortes
top_5_total_deaths = df_clean.loc[df_clean.date == '2023-01-13', ['location', 'total_deaths']].sort_values(by="total_deaths", ascending=False) [0:5]

fig, ax = plt.subplots(figsize=(15,10))
ax.set_title ('Paises com mais mortes  (Janeiro de 2023)', fontsize= 24)
ax.bar('location', 'total_deaths', data=top_5_total_deaths)

plt.tight_layout()
```

###### **Uma grande preocupação das organizações de saúde mundial é de como o virus está progredindo no número de casos e de mortes pelo mundo**
Para uma melhor análise, iremos plotar um gráfico de linha onde podemos verificar a demonstração da evolução do virus no mundo.

```python
# Evolução do total de casos do Covid-19
df_covid.loc[df_covid.location == "World", "total_deaths"].reset_index(drop=True).plot(figsize=(10,8));

```

Finalizando a nossa análise a nível mundial, vamos fazer uma rapida análise
entre a relação de total de mortes e o PIB per capita do paises. 


Para isso iremos plotar um gráfico de dispersão.

```python
# Gráfico de dispersão 
plt.figure(figsize = (16, 8))
sns.scatterplot(data=df_clean, x='gdp_per_capita', y='total_deaths')
plt.xlabel("PIB per Capita", size = 14)
plt.ylabel("Total de mortes", size = 14)
plt.show()
```

###### **Análise exploratória para o Brasil**
 Agora que ja vimos e podemos tirar algumas conclusões sobre o Covid-19 no mundo, iremos analisar alguns dados relevantes sobre o virus no Brasil.
 * Identificar quando foi registrada a primeira morte no Brasil
 * Identificar quantos dias demorou para registrarmos a primeira morte, considerando o primeiro caso de COVID-19 registrado.
 * Plotar gráficos de linha em escalas linear e logarítmica para os casos totais e mortes totais.

```python
# Criando Dataframe para o Brasil
df_brasil = df_clean.loc[df_clean.location == "Brazil"].copy()

df_brasil.head()
```

Criado o nosso DataFrame para o Brasil, vamos buscar a data mais recente para os dados do Brasil


```python
#Tamanho do Datafram
df_brasil.shape
```

```python
# Verificando a data mais recente de dados para o Brasil
df_brasil.date.max()
```

```python
# verificando data mais antiga para dados do Brasil
df_brasil.date.min()
```

Verificando a data mais recente e a data mais antiga de dados sobre o Brasil, notamos que existem dados de  **26/02/2020 até a data mais atual de 30/01/23**.

Feito isso, iremos identifica quando foi registrado o primeiro caso  e a primeira morte no Brasil.

```python
# Primeira caso de Covid-19 no Brasil
df_brasil.loc[df_brasil['new_cases'] != 0.0].head(1)
```

O primeiro caso de Covid-19 no Brasil foi registrado em **26/02/2020**

```python
# Primeira morte registrada no Brasil
df_brasil.loc[df_brasil['new_deaths'] == 1].head(1)
```

Uma informação que não podemos deixar de analisar é quanto tempo levou desde o primeiro caso confirmado até a primeira morte confirmada.


```python
# Identificar quantos dias demorou para registrarmos a primeira morte, considerando o primeiro caso de COVID-19 registrado.
# calculando tempo desde o primeiro caso até a primeira morte
from datetime import date
delta = df_brasil.date.loc[31155] - df_brasil.date.loc[31135]
print('Se passaram {} dias entre o primeiro caso e a primeira morte no Brasil.'.format(delta.days))
```

Por fim, vamos plotar gráficos de linhas para compararmos a evolução dos casos e mortes no Brasil.

```python
# Plotar gráficos de linha em escalas linear e logarítmica para os casos totais e mortes totais.
x = list(range(1078))
z = df_brasil['total_cases']
z2 = df_brasil['total_deaths']

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(9, 7))
ax1.set_title('Casos no Brasil')
ax1.set_xlabel('Dias')
ax1.set_ylabel('Casos')
ax1.plot(x, z)

ax2.set_title('Casos no Brasil (Escala logarítmica)')
ax2.set_xlabel('Dias')
ax2.set_ylabel('Casos')
ax2.set_yscale('log')
ax2.plot(x, z)

ax3.set_title('Mortes no Brasil')
ax3.set_xlabel('Dias')
ax3.set_ylabel('Mortes')
ax3.plot(x, z2)

ax4.set_title('Mortes no Brasil (Escala logarítmica')
ax4.set_xlabel('Dias')
ax4.set_ylabel('Mortes')
ax4.set_yscale('log')
ax4.plot(x, z2)
plt.tight_layout()
```
